# Trivial Matrix Multiplication

```python
import matplotlib.pyplot as plt
import time
import numpy as np

def existence_check(A,B):
  num_of_cols_0 = len(A[0])

  is_all_A_cols_len_equal = True
  for i in range(1,len(A)):
    if num_of_cols_0 != len(A[i]):
      is_all_A_cols_len_equal = False
  
  is_all_B_cols_len_equal = True
  for i in range(1,len(A)):
    if num_of_cols_0 != len(A[i]):
      is_all_B_cols_len_equal = False
  
  is_A_cols_equal_to_B_lines = num_of_cols_0 == len(B)
  
  if is_all_A_cols_len_equal & is_all_B_cols_len_equal & is_A_cols_equal_to_B_lines:
    return True
  
  return False
  
def trivial_matrix_mul(A,B): 
  if(not existence_check(A,B)):
    raise Exception("Matrixes A and B are not multiple")
  
  size_B = len(B)

  result = [[0 for y in range(size_B)] for x in range( size_B )]

  for i in range(size_B):
    for j in range(size_B):
      for k in range(size_B):
        result[i][j]+= A[i][k] * A[k][j]

simulations = 20
variants = 100
averages = []
n_vector = []

for j in range(2,variants + 1):

  variant =[[np.random.randint(-9999,9999) for y in range(j)] for x in range(j)]
  average_time = 0

  for i in range(simulations):

    initial_time = time.perf_counter_ns()

    trivial_matrix_mul(variant,variant)

    result_time = time.perf_counter_ns() - initial_time

    average_time += result_time

  averages.append(average_time/simulations)
  n_vector.append(j)

plt.title("Análise de tempo de execução")
plt.plot(n_vector, averages)
plt.xlabel("Tamanho da entrada N")
plt.ylabel("Tempo médio (nanosegundos)")
plt.show()
```
